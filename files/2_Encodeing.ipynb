```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
```

```python
Airbnb_data = pd.read_csv('Airbnb_data_V2.csv')
```

```python
Airbnb_data
```

```python
y = Airbnb_data['Price']

# Set the other columns as the independent variables, excluding 'City' and 'Room type'
X = Airbnb_data[['Person Capacity', 
                 'Multiple Rooms', 'Business', 'Cleanliness Rating', 'Guest Satisfaction', 'Bedrooms', 
                 'City Center (km)', 'Metro Distance (km)',
                 'Normalised Attraction Index', 'Normalised Restraunt Index']]



# Add a column of 1's to include an intercept in the model
X = sm.add_constant(X)

# Encode the 'Day' variable using binary encoding
encoded_data = pd.get_dummies(Airbnb_data['Day'], drop_first=True)

# Combine the encoded 'Day' variable with the existing independent variables
X = pd.concat([X, encoded_data], axis=1)

# Instantiate and fit the model
model = sm.OLS(y, X)
results = model.fit()

# Print the model summary
print(results.summary())
```
