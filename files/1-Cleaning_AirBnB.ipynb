```python
# Create the scatter plot:
plt.scatter(Airbnb_data['City Center (km)'], Airbnb_data['Price'])

# Labels for both axis:
plt.xlabel('City Center (km)')
plt.ylabel('Price')

# Title for the plot:
plt.title('Scatterplot of Price vs City Center (km))')
plt.show()
```

```python
# Calculate the average price for each room type:
average_prices = Airbnb_data.groupby('Room Type')['Price'].mean()

# Creating Bar Plot:
plt.bar(average_prices.index, average_prices)

# Labels:
plt.xlabel('Room Type')
plt.ylabel('Price')

# Plot Title:
plt.title('Average Prices by Room Type')
plt.show()
```

I found that the "room type" column is not very informative when assessing correlations, especially for "Entire Home/Apt". Therefore, I have decided to add a new column specifically for "Entire Home/Apt".

```python
Airbnb_data['Entire home/apt'] = Airbnb_data['Room Type'].apply(lambda x: True if x == 'Entire home/apt' else False)

cols = list(Airbnb_data.columns)

cols.insert(cols.index('Shared Room') - 1, cols.pop(cols.index('Entire home/apt')))

Airbnb_data = Airbnb_data[cols]

Airbnb_data
```

```python
# Select only numeric columns for correlation
numeric_data = Airbnb_data.select_dtypes(include=[np.number])

# Compute the correlation matrix on numeric data
corr_matrix = numeric_data.corr()

# Proceed with the heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
